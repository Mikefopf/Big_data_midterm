```python
import pandas as pd
from pandas import merge

from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
pd.set_option('display.max_columns', None)
```

```python
boarding_data = pd.read_csv("preprocessed/BoardingData.csv")
sirena_data = pd.read_csv("preprocessed/SirenaExport.csv")
flyer_forum_data = pd.read_csv("preprocessed/FrequentFlyerForumFlights.csv")
```

```python
boarding_data.shape
sirena_data.shape

joined_by_ticket = merge(boarding_data.iloc[:, :], sirena_data.iloc[:, :], how="inner",
               on=["TicketNumber"], suffixes=("_df1", "_df2"))
joined_by_ticket.shape
joined_by_ticket

joined_by_booking_code = merge(boarding_data.iloc[:, :], sirena_data.iloc[:, :], how="inner",
               on=["BookingCode"], suffixes=("_df1", "_df2"))
joined_by_booking_code.shape
joined_by_booking_code

joined_by_passport = merge(boarding_data.iloc[:, :], sirena_data.iloc[:, :], how="inner",
               on=["FlightNumber", "DepartureDate", "PassengerDocument"], suffixes=("_df1", "_df2"))
joined_by_passport.shape
joined_by_passport

joined_by_name = merge(boarding_data.iloc[:, :], sirena_data.iloc[:, :], how="inner",
               on=["FlightNumber", "DepartureDate", "PassengerFirstName", "PassengerLastName"], suffixes=("_df1", "_df2"))
joined_by_name.shape
joined_by_name

joined = joined_by_name
```

```python
joined_by_name = merge(joined.iloc[:, :], flyer_forum_data.iloc[:, :], how="inner",
               on=["FlightNumber", "DepartureDate", "PassengerFirstName", "PassengerLastName"], suffixes=("_df1", "_df2"))
joined_by_name.shape
joined_by_name
```
